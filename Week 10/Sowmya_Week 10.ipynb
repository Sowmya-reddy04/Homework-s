```python
import matplotlib.pyplot as plt
import numpy as np

# Define the steps and their corresponding metrics
steps = ['Dummy', 'KNN-1', 'KNN-2', 'KNN-3-1', 'KNN-3-2']
accuracy = [0.4767, 0.8983, 0.9717, 0.8983, 0.9717]
recall = [0.5075, 0.895, 0.963, 0.721, 0.965]
precision = [0.5327, 0.9045, 0.9061, 1.0, 0.8873]

# Set the x-axis positions
x = np.arange(len(steps))

# Create the plot
plt.figure(figsize=(10, 6))

# Plot accuracy, recall, and precision
plt.plot(x, accuracy, marker='o', label='Accuracy', color='blue')
plt.plot(x, recall, marker='o', label='Recall', color='orange')
plt.plot(x, precision, marker='o', label='Precision', color='green')

# Add labels and title
plt.xticks(x, steps)
plt.xlabel('Model Steps')
plt.ylabel('Performance Metrics')
plt.title('Batch Model Performance Metrics')
plt.ylim(0, 1)  # Set y-axis limits to [0, 1]
plt.grid()
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
