```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
from time import time
import numpy as np

dfdata = pd.read_csv('dfdata.csv')
dfdata.head()
```

```python
X = dfdata.drop('outcome', axis=1)
y = dfdata['outcome']

# Dataset sizes to test
sizes = [100, 1000, 10000, 100000, 1000000]

# Store results
results = []

for size in sizes:
    if len(dfdata) < size:
        print(f"Skipping size {size} - not enough data")
        continue

    # Sample data
    X_sample = X.iloc[:size]
    y_sample = y.iloc[:size]

    model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')

    # 5-fold stratified cross-validation
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    start_time = time()
    scores = cross_val_score(model, X_sample, y_sample, cv=cv, scoring='accuracy')
    elapsed = time() - start_time

    accuracy = np.mean(scores)
    results.append({
        'DatasetSize': size,
        'Accuracy': accuracy,
        'TimeTaken_sec': elapsed
    })

    print(f"Size: {size} | Accuracy: {accuracy:.4f} | Time: {elapsed:.2f} sec")

# Convert results to DataFrame
results_df = pd.DataFrame(results)
print(results_df)
```
